```python
import openai

client = openai.OpenAI()

PROMPT = """
내 시스템에 아래와 같은 함수들이 있어
`get_weather`
`get_currency`
`get_news`

이 함수들은 모두 나라의 이름을 인자로 받아. 예를들어 (get_weather('한국'))
이제 내가 실행해야 할 함수를 알려줘.
다른 사항은 언급하지 않고 함수 이름만 알려줘.

내가 물어볼 질문은
스페인의 날씨는 어떤가요? 
"""

response  = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": PROMPT},
    ],
)

response = response.choices[0].message.content
print(response)
```
