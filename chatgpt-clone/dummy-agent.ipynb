```python
from agents import Agent, Runner, function_tool


@function_tool
def get_weather():
    return "30"


agents = Agent(
    name = "Assistant Agent",
    instructions = "You are a helpful assistant. Use tools when needed to answer questions. in Korean",
    tools = [get_weather]
)

stream = Runner.run_streamed(agents, "안녕 날씨는 좀 어때?")

async for chunk in stream.stream_events():
    if chunk.type=="raw_response_event":
        continue
    elif chunk.type=="agent_updated_stream_event":
        print("Agent updated: ", chunk.new_agent)

    print("="*20)
```
