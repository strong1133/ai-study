```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model="deepseek-r1:1.5b")

llm.invoke("who am i")

```

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate(
    template="Where is the capital of the {country}?",
    input_variables=["country"],
)

prompt = prompt_template.invoke({"country": "korea"})
llm.invoke(prompt)
```

```python
from langchain_core.messages import HumanMessage

llm.invoke([HumanMessage(content="Where is the capital of the France")])
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

message_list = [
    SystemMessage(content="You are a helpful assistant"),
    AIMessage(content="The capital of France is Seoul"),
    HumanMessage(content="Where is the capital of the France"),
]

llm.invoke(message_list)
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","You are a helpful assistant"),
    ("human","Where is the capital of the {country}?")
])

chat_prompt = chat_prompt_template.invoke({"country", "Italy"})

chat_prompt.messages

llm.invoke(chat_prompt)
```
